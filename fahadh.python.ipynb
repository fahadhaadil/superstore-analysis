```python
top_states = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(5)
print(top_states)
```

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(category_profit)
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print(region_sales)
```

```python
sub_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
print(sub_profit)
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar', title='Sales by Category')
```

```python
df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum().plot(kind='line', title='Monthly Sales Trend')
```

```python
df.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%', title='Region-wise Sales')
```

```python
df.groupby('Sub-Category')['Profit'].sum().plot(kind='bar', title='Profit by Sub-Category')
```

Insight Summary

1.Technology category contributes the highest revenue.

Among all categories, Technology has the largest total sales, followed by Furniture and Office Supplies.

2.Western region shows the highest sales growth.

Sales in the West region are higher than in other regions, indicating strong market performance.

3.December and November are peak months for sales.

The monthly sales trend shows significant growth during these months, possibly due to holiday season demand.

4.Phones and Copiers are the most profitable sub-categories.

Certain sub-categories generate higher profit despite similar sales volumes, highlighting their contribution to overall profitability.

5.High sales do not always mean high profit.

Some items or categories have high sales but low profit, likely due to discounts or high costs.
